## Imports

### Arrays 

Un **array** es una estructura de datos que consiste en una coleccion de elementos(valores), cada uno identificado por almenos un indice o clave. Un array se elmacena de modo que la posicion de cada elemento se pueda calcular a paratir de su tupla de indice mediante su forma matematica .El tipo mas del array lineal tambien llamado array unidimencional

En Numpy:
* Cada dimencion se denomina **axis**.
* El numero de dimenciomnes se denomina **rank**
* La lista de dimenciones con su corresponndiente longitud se denomina **shape**
* El numero total de elementos (multiplicacion de la longitud de lass dimenciones ) se denomina **size**

```python
# Array cuyos valores son todos 0
a= np.zeros((2, 4))
```

```python
a
```

**a** es un array :

* Con dos **axis**, la parimera de longitud 2 y el segundo de longitud 4
* con un **rank** igual a 2
* con un **shape** igual (2,4) 
* con un **size** igual a 8

### Acceso a los elementos de un array 

#### Array Unidimencional

```python
# crar un array unidimencional 
array_uni = np.array([1, 3, 5, 7, 9, 11])
print("Shape: ",array_uni.shape)
print ("Array_uni: ", array_uni)
```

```python
# acceder al quinto elementos del array 
array_uni [4]
```

```python
# acceder al tercer y cuarto elemento del array
array_uni[2: 4]
```

```python
#acceder a elemento 0, 3 y 5 de array 
array_uni[[0, 3, 5]]
```

#### Array Multidimencional

```python
# crear un array Multidimencinal
array_multi= np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
print("Shape: ", array_multi)
print("Array_multi:  \n", array_multi )
      
```

```python
#acceder al cuarto elemento del array 
array_multi[0, 3]
```

```python
#acceder a la segunda fila del array
array_multi[1, :]
```

```python
# acceder al tercer elemnto de las 2 primeras filas del array 
array_multi[0: 2, 2]
```

#### Modificar un Array

```python
# Crear un array unidimencionla e inicializarlo con el rango de elementos del 0 - 27
array1=np.arange(28)
print("Shape: ", array1.shape)
print("Array1: \n", array1)
```

```python
# Cambiar las dimenciones del array y sus longitudes 
array1.shape = (7, 4)
print("Shape: ", array1.shape)
print("Array1: \n", array1)
```

```python
# El Ejemplo anterior devuelve un nuevo array que apunta los mismos datos 
#NOTA: las modificaciones de un array, modifican el otro array 
array2 = array1.reshape(4, 7)
print("Shape: ", array2.shape)
print("Array1: \n", array2)

print ("array1", array1.shape)
print("Array1: \n", array1)
```

```python
#Modificar el array devuelto 
array2 [0, 3] = 20
print("Shape: ", array1.shape)
print("Array1: \n", array1)

print("Shape: ", array2.shape)
print("Array1: \n", array2)
```

```python
#Regresar el array devolviendo el array de una sola dimencion
#NOTA: El aray apunta a los mismos datos
print("array1: ", array1.ravel())
```

### BROADCASTING 
Si se aplican operaciones aritmeticas sobre arrays que no tengas la misma forma **shape** Numpy aplica una propiedad que se denomina Broadcasting

```python
# Creacion de dos arryas unidimencionales 
array1 = np.arange(5)
array2 = np.array([3])

print("Shape Array 1:", array1.shape)
print("Array 1: ", array1)
print("\n")

print("Shape Array 2:", array2.shape)
print("Array 2:", array2)
```

```python
# Suma de ambos arreglos
array1 + array2
```

```python
# Crear un arreglo multidimencional y unidemencional
array1 = np.arange(6)
array1.shape = (2, 3)

array2 = np.arange(6, 18, 4)

print("Shape Array 1:", array1.shape)
print("Array 1: ", array1)

print("Shape Array 2:", array2.shape)
print("Array 2:", array2)
```

```python
# Suma de ambos arrays
array1 + array2
```
